# API data 2

```python
import requests

url = "https://odds.p.rapidapi.com/v4/sports/upcoming/odds"

querystring = {"regions":"us","oddsFormat":"decimal","markets":"h2h,spreads","dateFormat":"iso"}

headers = {
	"x-rapidapi-key": "56260415bbmsh46679066237bae8p13bf34jsn9e8329a5c06d",
	"x-rapidapi-host": "odds.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)

print(response.json())
```

```python
data_raw = json.loads(response.text)
data_raw
```

```python
df_odds = pd.json_normalize(data_raw)
df_odds.head()
```

As we can see the bookmakers is on json format, so we will separate it into another table. To make a relation to another table, we must have an ID. So, we wil add the ID for bookmakers

```python
# Extract unique bookmaker keys and map to IDs
unique_keys = {}
current_id = 1

def map_key_to_id(bookmaker_list):
    global current_id
    # Extract unique keys for this record
    keys = tuple(sorted(bm['key'] for bm in bookmaker_list))
    if keys not in unique_keys:
        unique_keys[keys] = f"B{current_id:03}"
        current_id += 1
    return unique_keys[keys]

# Apply the mapping
df_odds['bookmaker_id'] = df_odds['bookmakers'].apply(map_key_to_id)

# Drop the original 'bookmakers' column if no longer needed
df_odds = df_odds.drop(columns=['bookmakers'])
```

```python
df_odds.head()
```

And then we will make a bookmakers dataframe

```python
bookmakers_list = []

for i in range(len(data_raw)):
    bookmakers_list.append(data_raw[i]['bookmakers'])

bookmakers = [item for sublist in bookmakers_list for item in sublist]
bookmakers
```

```python
df_bookmakers = pd.json_normalize(bookmakers)
df_bookmakers.head()
```

As we can see the markets is also in json format. So we will separate it. Also, we have to add the bookmakers_id in df_bookmakers for joining with df_odds

```python
# Map `key` to IDs
unique_keys = {}
current_id = 1

def map_key_to_id(key):
    global current_id
    if key not in unique_keys:
        unique_keys[key] = f"B{current_id:03}"
        current_id += 1
    return unique_keys[key]

# Apply the mapping
df_bookmakers['bookmaker_id'] = df_bookmakers['key'].apply(map_key_to_id)

# Drop the original 'key' column if no longer needed
df_bookmakers = df_bookmakers.drop(columns=['key'])
```

```python
df_bookmakers.head()
```

```python
# Extract unique bookmaker representations and map them to IDs
unique_markets = {}
current_id = 1

def map_bookmaker_to_id(market_list):
    global current_id
    market_rept = str(sorted(market_list, key=lambda x: x['key']))
    if market_rept not in unique_markets:
        unique_markets[market_rept] = f"M{current_id:03}"
        current_id += 1
    return unique_markets[market_rept]

# Apply the mapping
df_bookmakers['market_id'] = df_bookmakers['markets'].apply(map_bookmaker_to_id)

df_bookmakers.head()
```

Make markets dataframe

```python
df_markets = pd.json_normalize(df_bookmakers['markets'])
df_markets.head()
```

There is many none values so we can drop the tables

```python
df_bookmakers = df_bookmakers.drop(columns=['markets','market_id'])
```

# API 3

```python
import requests

url = "https://odds.p.rapidapi.com/v4/sports/americanfootball_nfl/scores"

querystring = {"daysFrom":"3"}

headers = {
	"x-rapidapi-key": "56260415bbmsh46679066237bae8p13bf34jsn9e8329a5c06d",
	"x-rapidapi-host": "odds.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)

print(response.json())
```

```python
df_score = response.json()
df_score = pd.json_normalize(df_score)
df_score.head()
```

```python
df_score.isnull().sum()
```

```python
df_score = df_score.dropna()
```

Because the team is already on the dataframe, then we only extract the away_score and home_score from the 'scores' column

```python
# Extract away_score and home_score
df_score['away_score'] = df_score.apply(lambda x: next(int(score['score']) for score in x['scores'] if score['name'] == x['away_team']), axis=1)
df_score['home_score'] = df_score.apply(lambda x: next(int(score['score']) for score in x['scores'] if score['name'] == x['home_team']), axis=1)
df_score = df_score.drop(columns=['scores'])
```
